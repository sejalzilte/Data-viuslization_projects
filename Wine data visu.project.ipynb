#### Task1: Basic Distribution of Numerical Variables
##### Goal:Visualize the distribution of key numerical features.

```python
#Plotting histograms for all numerical columns
data.hist(bins=15, figsize=(15,12))
plt.tight_layout()
plt.show()
```

#### Task2:Correlation Heatmap
##### Goal:Visualize realtionships between numerical features.

```python
# Create a correlation heatmap
plt.figure(figsize=(12,8))
corr=data.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', cbar=True, square=True)
plt.title('Correlation Heatmap')
plt.show()
```

#### Task 3: Wine Quality vs Alcohol Content
##### Goal: Explore the relationship between wine quality and alcohol content.


```python
# Create a boxplot to visualize alcohol content by wine quality
plt.figure(figsize=(8,6))
sns.boxplot(x='quality', y='alcohol', hue='quality', data=data)
plt.title('Wine Quality vs Alcohol Content')
plt.show()
```

#### Task 4:Quality Distribution
##### Goal:Understand the distribution of wine quality scores.

```python
#Create a countplot for the wine quality distribution 
plt.figure(figsize=(8,6))
sns.countplot(x='quality', data=data, palette='viridis')
plt.title('Distribution of Wine Quality')
plt.show()
```

#### Task 5:Pairplot Analysis for Wine Quality
##### Goal:Explore relationships between wine quality and several key chemical features.

```python
# Display the column names
print(list(data.columns))          #List all column names
```

```python
#Update the feature list
selected_features=['alcohol', 'sulphates', 'citric acid', 'fixed acidity']
```

```python
# Set up the figure with sulphates: 2 rows and 4 columns for boxplot and histograms
plt.figure(figsize=(18,10))

#Loop through each feature and create a boxplot and histogram
for i,feature in enumerate(selected_features, 1):
    # Boxplot (subplot 1)
    plt.subplot(2,4,i)     #2 rows, 4 columns, feature position
    sns.boxplot(x=data[feature])
    plt.title(f'Boxplot of {feature}')


    # Histogram (subplot 2)
    plt.subplot(2,4,i+4)    #2 rows, 4 columns, feature position (second row)
    sns.histplot(data[feature], kde=True, bins=15, color='green')
    plt.title(f'Histogram of {feature}')

plt.tight_layout()
plt.show()
```

#### Task 6: Scatter Plot for Fixed Acidity vs Alcohol Content
##### Goal:Investigate the realtionship between acidity and alcohol content.

```python
# Scatter plot for fixed acidity vs alcohol content
plt.figure(figsize=(8,6))
sns.scatterplot(x='fixed acidity', y='alcohol', data=data, hue='quality',palette='viridis')
plt.title('Fixed Acidity vs Alcohol Content')
plt.show()

#Scatter plot for volatile acidity vs alcohol content
plt.figure(figsize=(8, 6))
sns.scatterplot(x='volatile acidity', y='alcohol', data=data, hue='quality',
                palette='viridis')
plt.title('Volatile Acidity vs Alcohol Content')
plt.show()
```

#### Task 7: Impact of Sulphar Dioxide on Wine Quality
##### Goal: Explore the impact of sulphar dioxide levels on wine quality.

```python
#Scatter plot for free sulphar dioxide vs quality
plt.figure(figsize=(8,6))
sns.scatterplot(x="free sulfur dioxide",y="quality",hue="quality",data=data,
                palette="viridis")
plt.title('Free Sulphar Dioxide vs Wine Quality')
plt.show()

#Scatter plot for total sulphar dioxide vs quality
plt.figure(figsize=(8,6))
sns.scatterplot(x='total sulfur dioxide',y='quality',hue='quality',data=data,
                palette='viridis')
plt.title('Total Sulphar Dioxide vs Wine Quality')
plt.show()
```

#### Task 8:Boxplot for Quality vs Various Features
##### Goal:Explore the impact of several features on wine quality.

```python
#Create boxplots for multiple features
features=['fixed acidity','residual sugar','sulphates','citric acid','density']
plt.figure(figsize=(16,10))

for i, feature in enumerate(features,1):
    plt.subplot(3,2,i)
    sns.boxplot(x='quality',y=feature, data=data,palette='viridis')
    plt.title(f'Quality vs {feature}')
plt.tight_layout()
plt.show()
```
